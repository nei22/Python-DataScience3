# Análise de Dados com Python

### Desafio:

Você trabalha em uma empresa de telecom e tem clientes de vários serviços diferentes, entre os principais: internet e telefone.

O problema é que, analisando o histórico dos clientes dos últimos anos, você percebeu que a empresa está com Churn de mais de 26% dos clientes.

Isso representa uma perda de milhões para a empresa.

O que a empresa precisa fazer para resolver isso?

```python
# - Definir os passos a serem seguidos para ANALIZAR DADOS
import pandas as pd #LIB DE ANALIZES
# - PASSO 1 - BASE DE DADOS - A) - IMPORTAR E LER DADOS,
table = pd.read_csv("telecom_users.csv")
# - PASSO 2 - BASE DE DADOS - B) - VISUALIZAR DADOS,
# display(table)

# - PASSO 3 - BASE DE DADOS - C) - CORRIGIR DADOS,
# - ELIMINAR COLUNA INUTIL

# axis = 0 -> line
# axis = 1 -> column
table = table.drop('Unnamed: 0', axis=1)
display(table)
```

```python
# - VALORES RECONHECIDOS DE FORMA ERRADA
# - object -> texto
# - int -> numeros inteiros
# - float -> numeros decimais
table['TotalGasto'] = pd.to_numeric(table['TotalGasto'], errors='coerce')
# - TRATAR VALORES VAZIOS
# excluir colunas vazios
table = table.dropna(how='all', axis=1)
#excluir as linhas vazias

table = table.dropna(how='any', axis=0)
display(table.info())
```

```python
# - PASSO 4 - BASE DE DADOS - D) - ANALIZE INICIAL DE DADOS.
print(table['Churn'].value_counts())
print(table['Churn'].value_counts(normalize=True).map('{:.1%}'.format))
```

```python
# - PASSO 5 - BASE DE DADOS - D) - ANALIZE DETALHADA DOS DADOS,
import plotly.express as px
# para edicoes graficas -> https://splotly.com/python/histogram/
for colunas in table.columns:
    grafic = px.histogram(table, x=colunas, color='Churn')
    grafic.show()
```
